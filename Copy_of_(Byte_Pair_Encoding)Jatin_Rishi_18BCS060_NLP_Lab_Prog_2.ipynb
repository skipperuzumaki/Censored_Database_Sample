```python
import collections
def read_file(file_name):
  my_file = open(f'/{file_name}', 'r').read()
  words = my_file.split(" ")
  print(f"Vocabulary Size =  {len(words)}")
  for item in words:
    print(item)
  return words
```

```python
def generate_tokenized_words(words):
  frequency_map = collections.defaultdict(int)

  for word in words:
      letters_split = [letter for letter in word]
      frequency_map[' '.join(letters_split)+ ' </w>'] += 1

  print("Words after adding unique end identifier : ")
  for item in frequency_map:
      print(f"{item} -> {frequency_map[item]}")
  return frequency_map
```

```python
def generate_word_pairs(frequency_map):
  pairs_frequency_map = collections.defaultdict(int)

  for item in frequency_map:
      letters = item.split()
      for idx in range(1,len(letters)):
          pairs_frequency_map[letters[idx-1], letters[idx]]+= frequency_map[item]

  for item in pairs_frequency_map:
      print(f"{item} -> {pairs_frequency_map[item]}",end="-------")
    
  print()

  return pairs_frequency_map
```

```python
import re
def merge_best_pairs(most_frequent_set,frequency_map):
    print(most_frequent_set)
    new_dict = {}
    filtered_set = re.escape(' '.join(most_frequent_set))
    pattern = re.compile(r'(?<!\S)' + filtered_set + r'(?!\S)')
    for word in frequency_map:
        matched_word = pattern.sub(''.join(most_frequent_set), word)
        new_dict[matched_word] = frequency_map[word]
    return new_dict
```

```python
def generate_word_tokens(frequency_map):
    character_frequency_map = collections.defaultdict(int)

    for item in frequency_map:
      letters_map = item.split(' ')
    for letter in letters_map:
      character_frequency_map[letter]+= frequency_map[item]

    for item in character_frequency_map:
      print(f"{item} -> {character_frequency_map[item]}",end="------")
    
    print()

    return character_frequency_map
```

```python
##Training Process
frequency_map = generate_tokenized_words(read_file('input.txt'))
for i in range(51):
    word_pairs = generate_word_pairs(frequency_map)
    best_matched_pair = max(word_pairs, key=word_pairs.get)
    print(f"ITERATION NO. {i}: ")
    frequency_map = merge_best_pairs(best_matched_pair,frequency_map)
    word_tokens = generate_word_tokens(frequency_map)

    for item in word_tokens:
        print(f"{item} -> {word_tokens[item]}")  
  
    print(len(word_tokens))
```

```python
## TESTING
frequency_map_2 = generate_tokenized_words(read_file('test_corpus.txt'))
frequency_map_prev = frequency_map_2
while(1):
    word_pairs = generate_word_pairs(frequency_map_2)
    best_matched_pair = max(word_pairs, key=word_pairs.get)
    frequency_map_2 = merge_best_pairs(best_matched_pair,frequency_map_2)
    if(frequency_map_2==frequency_map_prev):
      break
    else:
      frequency_map_prev = frequency_map_2
      word_tokens = generate_word_tokens(frequency_map_2)

    for item in word_tokens:
        print(f"{item} -> {word_tokens[item]}")  
  
    print(len(word_tokens))

print("Final Result")
print(word_tokens)
```
